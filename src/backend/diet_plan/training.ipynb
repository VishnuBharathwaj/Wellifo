```python
import tkinter as tk
from tkinter import messagebox, ttk
import os
import PyPDF2
import time
import datetime
import pyttsx3
import schedule
from plyer import notification
from threading import Thread

# Initialize text-to-speech engine
engine = pyttsx3.init()

# Global variables
schedule_enabled = False
selected_diet_pdf = ""
selected_exercise_pdf = ""

# Function to calculate BMI
def calculate_bmi(weight, height):
    bmi = weight / (height / 100) ** 2
    return round(bmi, 2)

# Function to load PDF content
def load_pdf_content(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            content = "".join([page.extract_text() for page in reader.pages])
            return content
    except FileNotFoundError:
        return "No plan available."

# Function to open a PDF file
def open_pdf(pdf_path):
    if os.path.exists(pdf_path):
        os.startfile(pdf_path)
    else:
        messagebox.showerror("Error", "File not found: " + pdf_path)

# Function to generate plan
def generate_plan():
    global selected_diet_pdf, selected_exercise_pdf
    
    # Get user inputs
    height = float(entry_height.get())
    weight = float(entry_weight.get())
    preference = combo_preference.get().strip().lower()
    if preference == "non-veg":
        preference = "non veg"
    
    # Calculate BMI
    bmi = calculate_bmi(weight, height)
    
    # Determine category
    category = 'underweight' if bmi < 18.5 else 'normal' if bmi <= 24.9 else 'overweight'
    
    # Paths for diet and exercise plans
    diet_paths = {
        'underweight': {
            'veg': r'C:\Users\HP\Downloads\diet-plan\Underweight - Veg Plan.pdf',
            'non veg': r'C:\Users\HP\Downloads\diet-plan\Non-Vegetarian Diet Plan - Underweight.pdf'
        },
        'normal': {
            'veg': r'C:\Users\HP\Downloads\diet-plan\veg-normal.pdf',
            'non veg': r'C:\Users\HP\Downloads\diet-plan\non-veg - normal.pdf'
        },
        'overweight': {
            'veg': r'C:\Users\HP\Downloads\diet-plan\overweight - ved.pdf',
            'non veg': r'C:\Users\HP\Downloads\diet-plan\overweight -  nonveg.pdf'
        }
    }

    exercise_paths = {
        'underweight': r'C:\Users\HP\Downloads\diet-plan\underweight excerise.pdf',
        'normal': r'C:\Users\HP\Downloads\diet-plan\normal - excerise.pdf',
        'overweight': r'C:\Users\HP\Downloads\diet-plan\overweight excerise.pdf'
    }
    
    selected_diet_pdf = diet_paths[category][preference]
    selected_exercise_pdf = exercise_paths[category]
    
    exercise_plan = load_pdf_content(selected_exercise_pdf)
    
    # Display in text widget
    text_output.delete(1.0, tk.END)
    text_output.insert(tk.END, f"BMI: {bmi}\nCategory: {category.capitalize()}\n\nExercise Plan:\n{exercise_plan}")

# Function for notifications
def notify(title, message):
    notification.notify(title=title, message=message, timeout=5)
    engine.say(message)
    engine.runAndWait()

# Function to enable schedule
def enable_schedule(schedule_label, frame, label):
    global schedule_enabled
    schedule_enabled = True
    notify("Great!", "Stay healthy and follow your schedule!")
    schedule_label.config(text=show_schedule())
    label.destroy()
    frame.destroy()
    Thread(target=schedule_runner, daemon=True).start()

# Function to run schedule
def schedule_runner():
    while True:
        schedule.run_pending()
        time.sleep(30)

# Function to display schedule
def show_schedule():
    return """
    Your Daily Health Schedule:
    - 06:30 AM: Wake Up
    - 07:00 AM: Morning Walk
    - 07:30 AM: Exercise
    - 08:00 AM: Breakfast
    - 10:30 AM: Mid-Morning Snack
    - 01:00 PM: Lunch
    - 04:00 PM: Evening Snack
    - 08:00 PM: Dinner
    - Hydration Reminder: Every 30 Minutes
    - 05:00 PM: Hobby Time
    - 10:00 PM: Sleep Time
    """

# Create GUI
def create_gui():
    root = tk.Tk()
    root.title("Health Assistant")
    root.geometry("500x600")
    
    tk.Label(root, text="Height (cm):").pack()
    global entry_height
    entry_height = tk.Entry(root)
    entry_height.pack()
    
    tk.Label(root, text="Weight (kg):").pack()
    global entry_weight
    entry_weight = tk.Entry(root)
    entry_weight.pack()
    
    tk.Label(root, text="Preference (Veg/Non-Veg):").pack()
    global combo_preference
    combo_preference = ttk.Combobox(root, values=["Veg", "Non-Veg"])
    combo_preference.pack()
    
    tk.Button(root, text="Generate Plan", command=generate_plan).pack()
    
    tk.Button(root, text="Open Diet PDF", command=lambda: open_pdf(selected_diet_pdf)).pack()
    tk.Button(root, text="Open Exercise PDF", command=lambda: open_pdf(selected_exercise_pdf)).pack()
    
    global text_output
    text_output = tk.Text(root, height=15, width=50)
    text_output.pack()
    
    schedule_label = tk.Label(root, text=show_schedule(), font=("Arial", 10), justify=tk.LEFT)
    schedule_label.pack(pady=10)
    
    frame = tk.Frame(root)
    frame.pack(pady=10)
    
    tk.Button(frame, text="Follow Schedule", command=lambda: enable_schedule(schedule_label, frame, None)).pack()
    
    root.mainloop()

# Scheduling reminders
schedule.every().day.at("06:30").do(lambda: notify("Wake Up Alert", "Time to wake up!"))
schedule.every().day.at("07:00").do(lambda: notify("Exercise Alert", "Time for morning walk!"))
schedule.every().day.at("07:30").do(lambda: notify("Exercise Alert", "Time for workout!"))
schedule.every().day.at("08:00").do(lambda: notify("Diet Alert", "Time for breakfast!"))
schedule.every(30).minutes.do(lambda: notify("Hydration Alert", "Drink water!"))
schedule.every().day.at("22:00").do(lambda: notify("Sleep Alert", "Time to sleep!"))

create_gui()
```
